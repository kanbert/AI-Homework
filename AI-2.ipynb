```python
#2
def person_is_seller():
    graph = dict()
    graph["you"] = ["bob", "claire", "alice"]
    graph["bob"] = ["peggy", "anuj"]
    graph["alice"] = ["peggy"]
    graph["claire"] = ["thom", "jonny"]
    graph["anuj"] = []
    graph["peggy"] = []
    graph["thom"] = []
    graph["jonny"] = []
    
    if graph is None:
        return
    nodeSet = set()
    stack = []
    nodeSet.add("you")
    stack.append("you")
    while len(stack) > 0:
        cur = stack.pop()
        for nexT in graph[cur]:
            if nexT not in nodeSet:
                stack.append(cur)
                stack.append(nexT)
                nodeSet.add(nexT)
                if nexT[-1] == 'm':
                    print(nexT)
person_is_seller()
```

```python
#3
import os

def dfsStack(path):    
    stack = []
    stack.append(path)
    while len(stack) > 0:
        filePath = stack.pop()
        fileList = os.listdir(filePath)
        for p in fileList:
            absFilePath = os.path.join(filePath, p)
            if os.path.isdir(absFilePath):
                stack.append(absFilePath)
            else:
                if p[-3:] == 'dll':
                    print(p)
                                
                                
path = 'C:\\'
dfsStack(path)
```

```python
import os

def dfsPath(path):
    if os.path.isdir(path):
        fileList = os.listdir(path)
        filePath = [os.path.join(path, file) for file in fileList]
        for path in filePath:
            dfsPath(path)
    else:
        if path[-3:] == 'dll':
            print(os.path.split(path)[1])
        return
                                            
                                
path = 'C:\\'
dfsPath(path)
```

```python
import os
from collections import deque
def bfsStack(path):    
    que = deque()
    pathlist = [path]
    que += pathlist
    researched = []
    while que:
        print(que)
        filepath = que.popleft()
        print(filepath)
        if filepath not in researched:
            researched.append(filepath)
            fileList = os.listdir(filepath)
            filePath = [os.path.join(path, filepath) for file in fileList]
            for fp in filePath:
                if os.path.isdir(fp):
                    pathlist = [fp]
                    que += pathlist
                else:
                    if fp[-3:] == 'dll':
                        print(os.path.split(path)[1])
                                
                                
path = 'C:\\'
bfsStack(path)
```
