```python
 class MinimaxAgent(MultiAgentSearchAgent):
    def getAction(self, gameState):
        self.num_agents = gameState.getNumAgents()
        a = gameState.getLegalActions(0)
        e = -float("inf")
        t = 0
        for N in a:
            if N == Directions.STOP:
                continue
            else:
                n = gameState.generateSuccessor(0, N)
                s = self.min_value(n, 1, 0)
                if s <= e:
                    continue
                else:
                    e = s
                    t = N

        return t

    def max_value(self, state, Z, h):
        if state.isWin() or state.isLose() or h == self.depth:
            return self.EFunction(state)
        else:
            U = -float("inf")
            A = state.getLegalActions(Z)
            for N in A:
                if N == Directions.STOP:
                    continue
                else:
                    n = state.generateSuccessor(Z, N)
                    r = self.min_value(n, Z + 1, h)
                    U = max(r, U)

            return U

    def min_value(self, state, Z, h):
        U = float("inf")
        if state.isWin() or state.isLose() or h == self.depth:
            return self.EFunction(state)
        if (Z + 1) % self.num_agents != 0:
            A = state.getLegalActions(Z)
            for N in A:
                if N == Directions.STOP:
                    continue
                else:
                    n = state.generateSuccessor(Z, N)
                    r = self.min_value(n, Z + 1, h)
                    U = min(r, U)
            return U
        else:
            A = state.getLegalActions(Z)
            for N in A:
                if N == Directions.STOP:
                    continue
                else:
                    n = state.generateSuccessor(Z, N)
                    r = self.max_value(n, 0, h + 1)
                    U = min(r, U)
            return U
```
